```python
import pandas as pd
df = pd.read_csv("NIPS_1987-2015.csv")
```

```python
# removed column 'Unnamed: 0' since it has string values
x = df.drop(columns = 'Unnamed: 0')
x
```

```python
from scipy.cluster.hierarchy import linkage, dendrogram

Z = linkage(x, 'ward')
```

```python
import matplotlib.pyplot as plt

%matplotlib inline

plt.figure(figsize=(25,25))
dendrogram(Z)
plt.show()
```
